```python
import matplotlib.pyplot as plt
```

```python
hours_studied=[10,9,2,15,10,16,11,16]
exam_scores=[95,80,10,50,45,98,35,93]
```

```python
plt.plot(hours_studied,exam_scores,marker='*',color='green',linestyle='-',mec='red')
plt.xlabel('Number of hours spent in studing')
plt.ylabel('Score in Exam')
plt.title('Effect of study hours on exam Performance')
plt.grid(True)
plt.show()
```

```python
plt.plot(hours_studied,exam_scores,marker='*',color=(255/255, 153/255, 51/255),linestyle='-',mec='blue')
plt.xlabel('Number of hours spent in studing')
plt.ylabel('Score in Exam')
plt.title('Effect of study hours on exam Performance')

#for annotation

for x,y in zip(hours_studied,exam_scores):
    plt.annotate(f'({x},{y})',(x,y))
plt.grid(True)
plt.show()
```

```python
# BarChart
plt.xlabel('Number of hours spent in studing')
plt.ylabel('Score in Exam')
plt.title('Effect of study hours on exam Performance')
plt.bar(hours_studied,exam_scores)
plt.grid()
```

```python
#Scatterplot
import numpy as np
plt.xlabel('Number of hours spent in studing')
plt.ylabel('Score in Exam')
plt.title('Effect of study hours on exam Performance')
colors=np.array([0,10,20,30,40,60,80,100])
plt.scatter(hours_studied,exam_scores,c=colors)
plt.colorbar()
plt.grid()
```
