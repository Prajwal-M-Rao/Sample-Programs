```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mtcars=pd.read_csv("mtcars.csv")
mtcars
```

```python
mpg_data=mtcars['mpg']
```

```python
plt.hist(mpg_data,bins=10,color='skyblue',edgecolor='black')
plt.xlabel('Miles per Gallon(mpg)')
plt.ylabel('Frequency')
plt.title('Frequency Distribtion of Miles per Gallon(mpg)')
plt.grid(True)
plt.show()
```
